```python

def plot_confusion_matrix(matrix):
    disp = ConfusionMatrixDisplay(confusion_matrix=matrix, display_labels=['CW','ACW','Other'])
    colours = ['#ffffcc','#a1dab4','#41b6c4','#2c7fb8','#253494']
    custom_cmap = LinearSegmentedColormap.from_list('custom_cmap', colours, N=20)
    disp.plot(cmap = custom_cmap,values_format=".3f",colorbar=False)
    for labels in disp.text_.ravel():
            labels.set_fontsize(26)
    ax = disp.ax_
    ax.set_xlabel('Predicted Labels',fontsize=14)
    ax.set_ylabel('True Labels',fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.tick_params(axis='both', which='minor', labelsize=10)
    plt.show()

gresnet18 = np.array([[0.754,0.001,0.246],
[0.001,0.720,0.279],
[0.016,0.017,0.967]])
gresnet18_flipeq = np.array([[0.289,0.281,0.430],
[0.298,0.257,0.445],
[0.010,0.009,0.981]])
plot_confusion_matrix(gresnet18)
plot_confusion_matrix(gresnet18_flipeq)
```
